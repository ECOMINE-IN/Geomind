```python
import pandas as pd
import numpy as np

master_df = pd.read_csv(
    "../data/processed/master_spectra_asd.csv"
)

print(master_df.shape)

master_df.head()
```

```python
print(master_df.info())
```

```python
print(master_df["mineral"].value_counts())
```

```python
print("Missing values:", master_df.isna().sum().sum())
```

```python
X = master_df.drop(
    columns=["mineral", "filename"]
)

y = master_df["mineral"]

print("X shape:", X.shape)
print("y shape:", y.shape)
```

```python
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()

y_encoded = label_encoder.fit_transform(y)

print("Number of classes:", len(label_encoder.classes_))

for mineral, code in zip(
    label_encoder.classes_[:10],
    range(10)
):
    print(f"{mineral} -> {code}")
```

```python
print(y_encoded[:20])
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_scaled = scaler.fit_transform(X)

print(X_scaled.shape)
```

```python
print("Mean:", np.mean(X_scaled))
print("Std :", np.std(X_scaled))
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X_scaled,
    y_encoded,
    test_size=0.2,
    random_state=42,
    stratify=y_encoded
)

print("Train:", X_train.shape)
print("Test :", X_test.shape)
```

```python
import numpy as np
```

```python
np.save("../data/processed/X_train.npy", X_train)
np.save("../data/processed/X_test.npy", X_test)

np.save("../data/processed/y_train.npy", y_train)
np.save("../data/processed/y_test.npy", y_test)
```
