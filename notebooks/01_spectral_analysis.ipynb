```python
import os

mineral_path = "../data/raw/usgs/usgs_splib07/ASCIIdata/ASCIIdata_splib07a/ChapterM_Minerals"

sample_file = os.path.join(
    mineral_path,
    "splib07a_Actinolite_HS116.1B_ASDFRb_AREF.txt"
)

with open(sample_file, "r") as f:
    for i in range(20):
        print(f.readline())
```

```python
wave_file = "../data/raw/usgs/usgs_splib07/ASCIIdata/ASCIIdata_splib07a/splib07a_Wavelengths_ASD_0.35-2.5_microns_2151_ch.txt"

wavelengths = []

with open(wave_file, "r") as f:

    next(f)  # Skip header

    for line in f:
        line = line.strip()

        if line:
            wavelengths.append(float(line))

print("Total wavelengths:", len(wavelengths))
print(wavelengths[:10])
```

```python
spectrum = []

with open(sample_file, "r") as f:

    next(f)  # Skip header

    for line in f:
        line = line.strip()

        if line:
            spectrum.append(float(line))

print("Total reflectance values:", len(spectrum))
print(spectrum[:10])
```

```python
import pandas as pd

df = pd.DataFrame({
    "Wavelength": wavelengths,
    "Reflectance": spectrum
})

df.head()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12,5))

plt.plot(
    df["Wavelength"],
    df["Reflectance"]
)

plt.xlabel("Wavelength (microns)")
plt.ylabel("Reflectance")
plt.title("Actinolite Spectral Signature")

plt.grid(True)

plt.show()
```

```python
import os

mineral_path = "../data/raw/usgs/usgs_splib07/ASCIIdata/ASCIIdata_splib07a/ChapterM_Minerals"

files = [f for f in os.listdir(mineral_path) if f.endswith(".txt")]

print("Total mineral spectra:", len(files))

for f in files[:10]:
    print(f)
```

```python
import pandas as pd

metadata = []

for file in files:

    parts = file.replace(".txt", "").split("_")

    mineral = parts[1]

    metadata.append({
        "filename": file,
        "mineral": mineral
    })

metadata_df = pd.DataFrame(metadata)

metadata_df.head()
```

```python
print("Unique minerals:", metadata_df["mineral"].nunique())

metadata_df["mineral"].value_counts().head(20)
```

```python
metadata_df["mineral"].value_counts().describe()
```

```python
metadata_df["mineral"].value_counts().tail(20)
```

```python
counts = metadata_df["mineral"].value_counts()

print(">=10 samples :", (counts >= 10).sum())
print(">=5 samples  :", (counts >= 5).sum())
print(">=3 samples  :", (counts >= 3).sum())
print(">=2 samples  :", (counts >= 2).sum())
```
